```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imgpath = "Red_example.png"
img = cv2.imread(imgpath)
```

```python
hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
RED_MIN = np.array([0, 120, 120],np.uint8)
RED_MAX = np.array([10, 255, 255],np.uint8)
red_light = cv2.inRange(hsv_img, RED_MIN, RED_MAX)
r = cv2.countNonZero(red_light)
r
```

```python
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
plt.imshow(red_light)
```

```python
imgpath = "Yellow_example.png"
img = cv2.imread(imgpath)
hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
YELLOW_MIN = np.array([40.0/360*255, 120, 120],np.uint8)
YELLOW_MAX = np.array([66.0/360*255, 255, 255],np.uint8)
yellow_color = cv2.inRange(hsv_img, YELLOW_MIN, YELLOW_MAX)
y = cv2.countNonZero(yellow_color)
y
```

```python
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
plt.imshow(yellow_color)
```

```python
imgpath = "Green_example.png"
img = cv2.imread(imgpath)
hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
GREEN_MIN = np.array([90.0/360*255, 100, 100],np.uint8)
GREEN_MAX = np.array([140.0/360*255, 255, 255],np.uint8)
green_color = cv2.inRange(hsv_img, GREEN_MIN, GREEN_MAX)
g = cv2.countNonZero(green_color)
g
```

```python
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```
